# This script is to test for llama3.2 3b instruct

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
```

### This is how to inference the model: 

```python
model_id = "meta-llama/Llama-3.2-3B-Instruct"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id,      device_map ="auto",)
```

```python
messages = [
    {"role": "system", "content": "You are an assistant for coding"},
    {"role": "user", "content": "Create a java function that sort a list. Give me the most efficient one"}
]

text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
)
print(text)
```

```python
model_inputs = tokenizer([text], return_tensors="pt").to("cuda")
print(model_inputs)
```

The generate method returns the input ids directly

```python
generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=8100,
    pad_token_id=tokenizer.eos_token_id,  # Ensure the model can generate the end-of-sequence token
    eos_token_id=tokenizer.eos_token_id # Define the end-of-sequence token
)

generated_ids = [
  output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]

response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
print(response)
```

Now let's get the whole output

```python
generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=300,
    pad_token_id=tokenizer.eos_token_id,  # Ensure the model can generate the end-of-sequence token
    eos_token_id=tokenizer.eos_token_id,  # Define the end-of-sequence token
)
#print(generated_ids)
response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
print(response)
```

The model() outputs the logits.

```python
generated_ids1 = model(
    **model_inputs,
    max_new_tokens=300
)

# Step 1: Extract logits from the output
logits = generated_ids1.logits

# Step 2: Get the token IDs by taking the argmax across the vocabulary dimension
token_ids = torch.argmax(logits, dim=-1)

print(token_ids)


#generated_ids = [
#    output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids1)
#]
#print(generated_ids1)

tokenizer.batch_decode(token_ids, skip_special_tokens=True)
print(tokenizer.batch_decode(token_ids, skip_special_tokens=True))
```

### Inference for LLM for software Model Completion

```python
from datasets import load_dataset
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
```

```python
checkpoint = "D:\\LLM\\thesisPractical\\fine_tuned_models\\metta_llama\\Llama-3.2-3B-Instruct-software-model_completion"
```

```python
instruction = "You are an AI assistant that specializes in UML model completion. Given an incomplete UML model represented in JSON format, output the missing portions of the model in JSON format."

def format_chat_template_to_print(input):
    row_json = [
        {"role": "system", "content": instruction},
        {"role": "user", "content": f'Here is the incomplete UML model:\n{input}'}
    ] 
    return tokenizer.apply_chat_template(row_json, tokenize=False, add_generation_prompt=True)

def format_chat_template(input):
    row_json = [
        {"role": "system", "content": instruction},
        {"role": "user", "content": f'Here is the incomplete UML model:\n{input}'}
    ] 
    return tokenizer.apply_chat_template(row_json, tokenize=True, return_tensors="pt", add_generation_prompt=True).to("cuda")
```

```python
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
)
model = AutoModelForCausalLM.from_pretrained(
    checkpoint,
    device_map ="auto",
        load_in_4bit=True
)
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
data = test_dataset[0]['input']
text_to_print = format_chat_template_to_print(data)
print(text_to_print)
input_ids = format_chat_template(data)
```
